```python
import numpy
import scipy.interpolate
import matplotlib.pyplot as plt
```

```python
data_x = [-5, 0, 1,3,4,6]
data_y = [-4, 2, -2,-4,0,4]
f = scipy.interpolate.Akima1DInterpolator(data_x, data_y)
x = numpy.linspace(data_x[0], data_x[-1], num=100)
plt.plot(data_x, data_y, 'o')
xn = x
print(xn[-2:])
plt.plot(xn+0.5, f(xn))
```
